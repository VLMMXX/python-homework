```python
# Import the pathlib and csv library
from pathlib import Path
import csv

# Set the file path
csvpath = Path('./resources/budget_data.csv')

# Open the input path as a file object
with open(csvpath, 'r') as csvfile:

# Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=',')
    
    # Go to the next row from the start of the file
    # (which is often the first row/header) 
    header = next(csvreader)
    months = 0
    total_profit_loss = 0
    current_month = 0
    previous_month = 0
    monthly_change = 0
    greatest_profit= 0
    revenue_change= 0
    profit_date = "" 
    greatest_loss = 0
    loss_date = ""

    for row in csvreader:
        # Print the row
        #print(row)
        #print(row[1])
        total_profit_loss = int(row[1])+ total_profit_loss
        months = months+1
        if (months >1):
            current_month = int(row[1])
            monthly_change = monthly_change + (current_month-previous_month)
            revenue_change= current_month - previous_month
            if greatest_profit < revenue_change:
                greatest_profit = revenue_change
                profit_date = row[0]
            if greatest_loss > revenue_change:
                greatest_loss = revenue_change
                loss_date = row[0]
            
            
        
        previous_month = int(row[1])
        
    print(f"Financial Analysis")
    print(f"-------------------------------")
    print(f"Total months: {months}")
    print(f"Total: ${total_profit_loss}")
    #print(round(monthly_change/(months-1),2))
    monthly_change = round(monthly_change/(months-1),2)
    print(f"Average change: ${monthly_change}")
    print(f"Greatest Increase in Profits: {profit_date} (${greatest_profit})")
    print(f"Greatest Decrease in Profits: {loss_date} (${greatest_loss})")

    output_path = Path('output.txt')
    # Open the output path as a file object
    with open(output_path, 'w') as text_file:
    
        print(f"Financial Analysis",file=text_file)
        print(f"-------------------------------",file=text_file)
        print(f"Total months: {months}",file=text_file)
        print(f"Total: ${total_profit_loss}",file=text_file)
        #print(round(monthly_change/(months-1),2))
        monthly_change = round(monthly_change/(months-1),2)
        print(f"Average change: ${monthly_change}",file=text_file)
        print(f"Greatest Increase in Profits: {profit_date} (${greatest_profit})",file=text_file)
        print(f"Greatest Decrease in Profits: {loss_date} (${greatest_loss})",file=text_file)
    

   

    



```
